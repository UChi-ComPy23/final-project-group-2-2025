```python
# Import necessary libraries and modules
import sys
import os

# Add the project root directory to the Python path
sys.path.append(os.path.dirname(os.getcwd()))

import matplotlib.pyplot as plt
from src.graph import Graph
from src.greedy import greedy_coloring
from examples.utils import random_graph, plot_colored_graph
```

```python
# Run a small experiment to see how the greedy algorithm performs on different graph sizes
# We'll test graphs with different numbers of vertices and measure runtime and colors used

# List of graph sizes to test (number of vertices)
ns = [5, 10, 15, 20, 25, 30, 35, 40]
times = []
num_colors_list = []

for n in ns:
    # Create a random graph with n vertices
    # p=0.3 means each pair of vertices has a 30% chance of being connected
    g = random_graph(n, p=0.3)
    
    # Run greedy algorithm with degree ordering (color vertices with more neighbors first)
    result = greedy_coloring(g, use_degree_order=True)
    
    # Record the results
    times.append(result.time_seconds)
    num_colors_list.append(result.num_colors)
    
    # If we found a valid coloring, visualize it
    if result.coloring:
        plot_colored_graph(g, result.coloring, title=f"Greedy Graph Coloring (n={n}, colors={result.num_colors})")

# Create a figure with two plots side by side
plt.figure(figsize=(12, 5))

# Plot 1: Show how runtime changes with graph size
plt.subplot(1, 2, 1)
plt.plot(ns, times, marker='o')
plt.xlabel("n (number of vertices)")
plt.ylabel("runtime (seconds)")
plt.title("Greedy Algorithm Runtime vs Graph Size")
plt.grid(True)

# Plot 2: Show how many colors are used as graph size increases
plt.subplot(1, 2, 2)
plt.plot(ns, num_colors_list, marker='o')
plt.xlabel("n (number of vertices)")
plt.ylabel("number of colors")
plt.title("Greedy Algorithm: Colors Used vs Graph Size")
plt.grid(True)

plt.tight_layout()
plt.show()
```
