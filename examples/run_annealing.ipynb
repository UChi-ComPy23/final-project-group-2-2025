```python
import sys
import os

# add project root to path
sys.path.append(os.path.dirname(os.getcwd()))

import matplotlib.pyplot as plt
from src.graph import Graph
from src.annealing import simulated_annealing
from examples.utils import random_graph, plot_colored_graph
```

```python
# small experiment
ns = [10, 20, 40, 80, 160, 320, 640, 1280]
times = []

for n in ns:
    g = random_graph(n, p=0.3)
    result = simulated_annealing(g, k=4)
    times.append(result.time_seconds)
    if result.coloring:
        plot_colored_graph(g, result.coloring, title="Graph Coloring (k=4)")

plt.plot(ns, times, marker='o')
plt.xlabel("n (number of vertices)")
plt.ylabel("runtime (seconds)")
plt.title("Simulated Annealing Runtime vs Graph Size")
plt.show()
```
